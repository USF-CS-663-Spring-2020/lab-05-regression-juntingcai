# 1.Toluca Dataset

```python
import pandas as pd

df = pd.read_csv('toluca.csv')
df
```

```python
df.plot(kind='scatter',  x = 'lotSize', y = 'workHours')
```

```python
import seaborn as sns
plot = sns.lmplot(y = 'workHours', x = 'lotSize', data = df)
```

```python
rl = sns.residplot('lotSize', 'workHours', data = df)
rl.set(xlabel = 'Lot Size', ylabel = 'Residuals')
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(df.drop('workHours', axis = 1), df.iloc[:,1])
lin_reg.coef_
```

```python
lin_reg.intercept_
```

# 2.Credit dataset

```python
df = pd.read_csv('Credit.csv')
df
```

### Predictor: Income

```python
df.plot(kind = 'scatter', x = 'Limit', y = 'Income')
```

```python
import seaborn as sns
plot = sns.lmplot(y = 'Income', x = 'Limit', data = df)
```

```python
rl = sns.residplot('Limit', 'Income', data = df)
rl.set(xlabel = 'Limit', ylabel = 'Residuals')
```

```python
lin_reg = LinearRegression()
lin_reg.fit(df[['Limit']], df[['Income']])
print("slop: ",  lin_reg.coef_)
print("intercept: ", lin_reg.intercept_)
```
